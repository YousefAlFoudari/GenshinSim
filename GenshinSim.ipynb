```python
#Rolling logic

#Figure out what to do based on current pity
def roll(fourStarPity, fiveStarPity):
    if fourStarPity >= 8 and fiveStarPity >= 88:
        item = random.choices(fourStarList)
        fourStarPity = 0
        fiveStarPity+=1
    elif fourStarPity < 9 and fiveStarPity < 89:
        item, fourStarPity, fiveStarPity = gamble(fourStarPity, fiveStarPity)
    elif fourStarPity >= 9:
        item = random.choices(fourStarList)
        fourStarPity = 0
        fiveStarPity+=1
    elif fiveStarPity >= 89:
        item = random.choices(fiveStarList)
        fiveStarPity = 0
        fourStarPity+=1
    return item[0], fourStarPity, fiveStarPity

#Find out which rarity you're getting and update pity
def gamble(fourStarPity, fiveStarPity): 
    #soft 5-Star pity
    if fiveStarPity < 75:
        rarity = random.choices(rarityList, weights = [0.6, 5.1, 94.3])
    elif fiveStarPity >= 75 and fiveStarPity < 90:
        rarity = random.choices(rarityList, weights = [32.4, 5.1, 94.3])     
    elif fiveStarPity == 90:
        rarity = "5-Star"  
    #Select item at random based on rarity
    if rarity[0] == "3-Star":
        item = random.choices(threeStarList)
        fourStarPity+=1
        fiveStarPity+=1
    elif rarity[0] == "4-Star":
        item = random.choices(fourStarList)
        fourStarPity = 0
        fiveStarPity+=1
    elif rarity[0] == "5-Star":
        item = random.choices(fiveStarList)
        fourStarPity+=1
        fiveStarPity = 0
    return item, fourStarPity, fiveStarPity
```

```python
#Helper methods

#Match string with object of class Item, i.e. match "Qiqi" with Qiqi
def match(result):
    for entry in completeList:
        if entry.Name == result:
            return entry

#recieves string input and ensures it's an item from the Standard Banner
def getName():
    name = input()
    result = match(name)
    if len(name) > 50:
        print ("Input is too long! Please try again.")
        return getName()
    elif result == None:
        print("Item not found! Please select an item in the Standard Wish Banner. (Capitals and punctuation are necessary!)")
        return getName()
    return result.Name

#recieves integer input with a defined maximum
def getNumber(max):
    numWanted = int(input())
    if numWanted > max:
        print("Max", max, "! Try again!")
        return getNumber()
    return numWanted

#asks user for either yes or no 
def getAns():
    answer = input()
    if answer != "yes" and answer != "Yes" and answer != "no" and answer != "No":
        print('Please type either "Yes" or "No"')
        return getAns()
    return answer

#prints disclaimer
def printDisclaimer(): #TODO: don't print the disclaimer each time
    print()
    print("""*Note: The calculated "price" of any 5-Star item is based on obtaining Jean 10,000 times, and averaging the amount of money spent per Jean. I did this 10 seperate times (i.e. 10 instances of 10,000 acquired Jeans), and averaged the 10 results to obtain the final value shown in the statistics. My test case was Razor for 4-Star items, and Cool Steel for 3-Star items. I believe these numbers won't change significantly were I to select a different test case, as all 5-Star items share the same probability (the same reasoning goes for 4-Star and 3-Star items.) You may interpret these results however you please. """)
    print("** #$124.99/P8080 * P1600 = $24.7504950495 dollars per 10-roll (for my PS store), and 1 roll = 160 primogems")
```

```python
#main method
def main():
    for item in completeList:
        item.NumberOwned = 0
    print("Are you rolling for a specific item?")
    answer = getAns()
    if answer=="no" or answer == "No":  #roll without objective
        routeA()
    else:   #roll for specific player
        routeB()
        printDisclaimer()
    print("Would you like to roll again?")
    answer = getAns()
    if answer == "yes" or answer == "Yes":
        for item in completeList:
            item.NumberOwned = 0
        main()
    else:
        print("Thanks for using GenshinSim!")
        
```

```python
#logic for rolling without objective
def routeA():
    print("Okay, how many times would you like to roll? (max: 1,000,000)")
    numrolls = getNumber(1000000)
    fourStarPity = 0
    fiveStarPity = 0
    rollcounter = numrolls
    rollsPerFive = 0
    fiveArr = []
    print("Rolling in progress...")
    while rollcounter > 0:
        intermediate, fourStarPity, fiveStarPity = roll(fourStarPity, fiveStarPity)
        final = match(intermediate)
        final.NumberOwned += 1
        rollcounter -= 1
        if final.Rarity == "5-Star":
            fiveArr.append(rollsPerFive)
            rollsPerFive = 0
        else:
            rollsPerFive += 1
        #print(final.Name, fourStarPity, fiveStarPity)
    print("Rolling complete!")
    printStatsA(numrolls)
    plotStats(fiveArr)

    
#prints stats for rolling without objective
def printStatsA(numrolls):
    print(numrolls, "total rolls.") #TODO: want to print 2 decimal points
    usdSpent = numrolls/10 * 24.7504950495
    print("$" , '%.2f'%usdSpent, "dollars spent") #$124.99/P8080 * P1600 = $24.7504950495 (based on my PS Store)
    print(numrolls * 160, "primogems spent")
    print()
    numFiveStars = 0
    numFourStars = 0
    numThreeStars = 0
    for item in completeList:
        if item.NumberOwned > 0 and item.Rarity == "5-Star":
            numFiveStars+=item.NumberOwned
        elif item.NumberOwned > 0 and item.Rarity == "4-Star":
            numFourStars+=item.NumberOwned
        elif item.NumberOwned > 0 and item.Rarity == "3-Star":
            numThreeStars+=item.NumberOwned
    print()
    print("5-Star items obtained:" , numFiveStars)
    print("4-Star items obtained:" , numFourStars)
    print("3-Star items obtained:" , numThreeStars)
    print()
    print("Items recieved:")
    print()
    for item in completeList:
        if item.NumberOwned >= 1 and item.Rarity != "3-Star":
            printItem(item)
```

```python
#logic for rolling with objective
def routeB():
    print("Okay, how it works is you input which item you'd like to roll for, and how many of it you'd like.")
    print("""Please enter the name of your desired item exactly as it appears on the Standard Banner (i.e. Amos' Bow or Primordial Jade Winged-Spear). Capitals and punctuation are necessary.""")
    print()
    print("So, which item would you like?")
    itemWanted = getName()
    print("How many", itemWanted, "would you like? (max: 10,000)") #TODO: fix comma error
    numWanted = getNumber(10000)
    fourStarPity = 0
    fiveStarPity = 0
    totalRolls = 0
    rollsPerWanted = 0
    rollsPerFive = 0
    wantedArr = []
    fiveArr = []
    itemWanted = match(itemWanted)
    print("Rolling in progress...")
    while itemWanted.NumberOwned < numWanted:
        intermediate, fourStarPity, fiveStarPity = roll(fourStarPity, fiveStarPity)
        final = match(intermediate)
        final.NumberOwned += 1
        totalRolls += 1
        if final.Name == itemWanted.Name:
            wantedArr.append(rollsPerWanted)
            rollsPerWanted = 0
        else:
            rollsPerWanted += 1
        if final.Rarity == "5-Star":
            fiveArr.append(rollsPerFive)
            rollsPerFive = 0
        else:
            rollsPerFive += 1
    print("Rolling complete!")
    printStatsB(numrolls, numWanted, itemWanted, rollArr)
    plotStats(fiveArr)
    
    #prints statistics for specific player
def printStatsB(numrolls, numWanted, itemWanted, rollArr, fiveArr):
    print()
    print(numrolls, "total rolls")
    usdSpent = numrolls/10 * 24.7504950495
    print("$" , '%.2f'%usdSpent, "dollars spent**") #$124.99/P8080 * P1600 = $24.7504950495 (based on my PS Store)
    print(numrolls * 160, "primogems spent**")
    avgUsdSpent = usdSpent/numWanted
    print("Average: $", '%.2f'%avgUsdSpent, "dollars spent per", itemWanted.Name)
    print("Average:", '%.2f'%(numrolls*160/numWanted), "primogems spent per", itemWanted.Name)
    numFiveStars = 0
    numFourStars = 0
    numThreeStars = 0
    for item in completeList:
        if item.NumberOwned > 0 and item.Rarity == "5-Star":
            numFiveStars+=item.NumberOwned
        elif item.NumberOwned > 0 and item.Rarity == "4-Star":
            numFourStars+=item.NumberOwned
        elif item.NumberOwned > 0 and item.Rarity == "3-Star":
            numThreeStars+=item.NumberOwned
    print()
    print("5-Star items obtained:" , numFiveStars)
    print("4-Star items obtained:" , numFourStars)
    print("3-Star items obtained:" , numThreeStars)
    print()
    print("Maximum rolls to get", itemWanted.Name, ":", max(rollArr))
    print("Minimum rolls to get", itemWanted.Name, ":", min(rollArr))
    print()
    calAvgUSD, calAvgPrimo = calAvg(itemWanted)
    print(itemWanted.Name, 'has a calculated average "price" of $', calAvgUSD, "or", calAvgPrimo, "Primogems*")
    print()

#prints pre-calculated average based on item rarity 
def calAvg(item):
    if item.Rarity == "5-Star":
        calAvgUSD = 2674.13
        calAvgPrimo = 171144
    elif item.Rarity == "4-Star":
        calAvgUSD = 560.85
        calAvgPrimo = 35894
    elif item.Rarity == "3-Star":
        calAvgUSD = 37.55695
        calAvgPrimo = 2403
    return calAvgUSD, calAvgPrimo
```

```python
def plotStats(fiveArr):    
    count = 0
    countArr = [0] * 91 #TODO: make sure no 5* recieved at 90+
    for i in fiveArr:
        countArr[i] += 1
    
    # x axis values 
    x = range(0, 91)
    # corresponding y axis values 
    y = countArr

    # plotting the points  
    plt.plot(x, y) 

    # naming the x axis 
    plt.xlabel('Number of rolls') 
    # naming the y axis 
    plt.ylabel('Number of 5-Stars') 

    # giving a title to my graph 
    plt.title('') 

    # function to show the plot 
    plt.show() 
    
    #print(countArr[90])
```
